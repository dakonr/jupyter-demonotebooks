```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("Task_3_data.xlsx")
```

```python
data
```

```python
x = data.Newsletterabonnenten.values
y = data.Besucher.values
```

```python
x = x.reshape(len(x), 1)
y = y.reshape(len(y), 1)
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
r_sq = model.score(x, y)
intercept = model.intercept_
slope = model.coef_
print(f"intercept: {intercept}")
print(f"slope: {slope}")
print(f"R² {r_sq}")
```

```python
# plot
sns.set()
plt.scatter(x, y, alpha=0.5)
plt.xlabel("Newsletterabonnenten")
plt.ylabel("Besucher auf der Webseite")
t = (min(x), max(x))
plt.plot(t, model.predict(t), "-r")
ax = plt.gca()
plt.text(
    0.95,
    0.05,
    "R² = " + str(round(r_sq, 2)),
    horizontalalignment="right",
    verticalalignment="center",
    transform=ax.transAxes,
)
plt.show()
```

```python
p = sns.regplot(x="Newsletterabonnenten", y="Besucher", data=data, ci=None)
p.set_ylabel("Besucher auf der Webseite")
p.plot()
```
