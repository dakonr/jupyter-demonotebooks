```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Task_1_data.csv", sep=";", index_col=False, header=0)
```

```python
data
```

```python
x = data.Werbeetat.values
y = data.Zufriedenheit.values
```

```python
x = x.reshape(len(x), 1)
y = y.reshape(len(y), 1)
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
r_sq = model.score(x, y)
intercept = model.intercept_
slope = model.coef_
print(f"intercept: {intercept}")
print(f"slope: {slope}")
print(f"coeff {r_sq}")
```

```python
# plot
sns.set()
plt.scatter(x, y, alpha=0.5)
plt.xlabel("Werbeetat")
plt.ylabel("Zufriedenheit")
t = (min(x), max(x))
plt.plot(t, model.predict(t), "-r")
ax = plt.gca()
plt.text(
    0.95,
    0.05,
    "R² = " + str(round(r_sq, 2)),
    horizontalalignment="right",
    verticalalignment="center",
    transform=ax.transAxes,
)
plt.show()
```
