```python
CHESSBOARD = 8 * 8
MULTIPLY_BY = 2
RICE_ON_FIRST_FIELD = 1

def multiply_step(field):
    rice = RICE_ON_FIRST_FIELD * (2**field)
    if field == 0:
        if rice == 1:
            print(f'Das Schachbrett hat {CHESSBOARD} Felder und auf dem ersten Feld ist ein Reiskorn.\n',
                  f'bei jedem Feldwechsel wird die Reiskornanzahl mit {MULTIPLY_BY} multipiziert')
        if rice != 1:
            print(f'Das Schachbrett hat {CHESSBOARD} Felder und auf dem ersten Feld sind {rice} Reiskörner.\n',
                  f'bei jedem Feldwechsel wird die Reiskornanzahl mit {MULTIPLY_BY} multipiziert')
    else:
        print(f'auf dem Feld {field + 1} liegen {rice} Körner')
    return rice

print(f'Damit liegen',
      sum(map(multiply_step, range(CHESSBOARD))),
      'Reiskörner auf dem Schachbrett.')
```
