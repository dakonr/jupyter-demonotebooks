# Pi berechnen und darstellen - statistisch bestimmen
Aufgabe ist es, Ï€ (die Kreiszahl) zu berechenen anhand eines Einheitskreises und dieses grafisch darzustellen.

Wiki: https://de.wikipedia.org/wiki/Kreiszahl
Monte-Carlo-Algorithmus
Verwendete (externe) Module:
- numpy
- matplotlib

```python
%matplotlib inline
```

```python
from numpy import sqrt, linspace
from numpy.random import uniform

import matplotlib.pyplot as plt

from IPython.display import display_latex
```

```python
xs = linspace(0, 1, 1000)
ys = sqrt(1 - xs **2)

fig = plt.figure(figsize=(10, 10), dpi=100)
ax = fig.add_subplot(111)
plt.xlim(0, 1)
plt.ylim(0, 1)
ax.set_aspect('equal', adjustable='box')

SIZE = 1000000
pts = uniform(0, 1, size=(SIZE, 2))
inside = pts[:,0] ** 2 + pts[:,1] ** 2 < 1

plt.plot(xs, ys, color='blue')
plt.scatter(pts[inside][:,0], pts[inside][:,1], color='blue', s=1)
plt.scatter(pts[~inside][:,0], pts[~inside][:,1], color='red', s=1)

pi = sum(inside) / SIZE * 4

display_latex(r'$$ \pi \approx {} $$'.format(pi), raw=True)
plt.show()
```
